```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('/content/Motor_Vehicle_Collisions_-_Crashes.csv')

# Display the first few rows
print(df.head())

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Fill missing values if necessary (Example: Fill missing boroughs with 'Unknown')
df['BOROUGH'] = df['BOROUGH'].fillna('Unknown')

# Convert date and time columns to datetime objects
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'])

# Convert 'CRASH TIME' to datetime, handling errors
# errors='coerce' will set invalid values to NaT (Not a Time)
df['CRASH TIME'] = pd.to_datetime(df['CRASH TIME'], format='%H:%M', errors='coerce').dt.time

# If you expect AM/PM indicators, use this instead:
# df['CRASH TIME'] = pd.to_datetime(df['CRASH TIME'], format='%I:%M %p', errors='coerce').dt.time

# Check data types
print(df.dtypes)

# Fill missing numeric values with 0 (Example: Number of persons injured)
df.fillna(0, inplace=True)

# Remove any duplicate rows if present
df.drop_duplicates(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score, f1_score, recall_score
import pandas as pd

# Select features and target variable
features = df[['BOROUGH', 'LATITUDE', 'LONGITUDE', 'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2']]
target = df['NUMBER OF PERSONS INJURED']

# Encode categorical variables
features = pd.get_dummies(features)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Regressor model
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred_tree = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred_tree)
print('Mean Squared Error:', mse)

# Convert predicted values to integer
y_pred_tree = y_pred_tree.round().astype(int)

print("Decision tree")
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_tree)
print('Accuracy:', accuracy)

# Calculate precision
precision = precision_score(y_test, y_pred_tree, average='macro')
print('Precision:', precision)

# Calculate F1-score
f1 = f1_score(y_test, y_pred_tree, average='macro')
print('F1-score:', f1)

# Calculate recall
recall = recall_score(y_test, y_pred_tree, average='macro')
print('Recall:', recall)

# Feature importance
importance = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X_train.columns, 'Importance': importance}).sort_values(by='Importance', ascending=False)
print(feature_importance)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score, f1_score, recall_score
import pandas as pd


# Select features and target variable
features = df[['BOROUGH', 'LATITUDE', 'LONGITUDE', 'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2']]
target = df['NUMBER OF PERSONS INJURED']

# Encode categorical variables
features = pd.get_dummies(features)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Initialize and train the AdaBoost Regressor model
model = AdaBoostRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred_ada = model.predict(X_test)

print("AdaBoost")
# Evaluate the model
mse = mean_squared_error(y_test, y_pred_ada)
print('Mean Squared Error:', mse)

# Convert predicted values to integer
y_pred_ada = y_pred_ada.round().astype(int)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_ada)
print('Accuracy:', accuracy)

# Calculate precision
precision = precision_score(y_test, y_pred_ada, average='macro')
print('Precision:', precision)

# Calculate F1-score
f1 = f1_score(y_test, y_pred_ada, average='macro')
print('F1-score:', f1)

# Calculate recall
recall = recall_score(y_test, y_pred_ada, average='macro')
print('Recall:', recall)

# Feature importance
importance = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X_train.columns, 'Importance': importance}).sort_values(by='Importance', ascending=False)
print(feature_importance)
```
